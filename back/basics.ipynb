```python
def get_data_for_ticker(ticker):
    stock = yf.Ticker(ticker)
    
    info = stock.info
    data = {}

    data['Symbol'] = info.get('symbol', '')
    data['Name'] = info.get('longName', '')
    data['Industry'] = info.get('industry', '')
    data['EPS (fwd)'] = info.get('forwardEps', np.nan)
    data['P/E (fwd)'] = info.get('forwardPE', np.nan)
    data['PEG'] = info.get('pegRatio', np.nan)
    data['FCFY'] = info.get('freeCashflow', np.nan)
    data['PB'] = info.get('priceToBook', np.nan)
    data['ROE'] = info.get('returnOnEquity', np.nan)
    data['P/S (trail)'] = info.get('priceToSalesTrailing12Months', np.nan)
    data['DPR'] = info.get('dividendPayoutRatio', np.nan)
    data['DY'] = info.get('dividendYield', np.nan)
    data['CR'] = info.get('currentRatio', np.nan)
    data['Beta'] = info.get('beta', np.nan)
    data['Price'] = info.get('regularMarketPrice', np.nan)
    data['52w Low'] = info.get('fiftyTwoWeekLow', np.nan)
    data['52w High'] = info.get('fiftyTwoWeekHigh', np.nan)

    return data
```

```python
data = {
    'Symbol': [],
    'Name': [],
    'Industry': [],
    'EPS (fwd)': [],
    'P/E (fwd)': [],
    'PEG': [],
    'FCFY' : [],
    'PB': [],
    'ROE' : [],
    'P/S (trail)': [],
    'DPR' : [],
    'DY' : [],
    'CR' : [],
    'Beta': [],
    'Price': [],
    '52w Low': [],
    '52w High': []
}
# datas = []
bad_tickers = []
for symbol in symbols:
    # data = {}
    try:
        stock = yf.Ticker(symbol)
        info = stock.info
    except:
        bad_tickers.append(symbol)
        continue    
    
    data['Symbol'].append(info.get('symbol', ''))
    data['Name'].append(info.get('longName', ''))
    data['Industry'].append(info.get('industry', ''))
    data['EPS (fwd)'].append(info.get('forwardEps', np.nan))
    data['P/E (fwd)'].append(info.get('forwardPE', np.nan))
    data['PEG'].append(info.get('pegRatio', np.nan))
    data['FCFY'].append(info.get('freeCashflow', np.nan))
    data['PB'].append(info.get('priceToBook', np.nan))
    data['ROE'].append(info.get('returnOnEquity', np.nan))
    data['P/S (trail)'].append(info.get('priceToSalesTrailing12Months', np.nan))
    data['DPR'].append(info.get('dividendPayoutRatio', np.nan))
    data['DY'].append(info.get('dividendYield', np.nan))
    data['CR'].append(info.get('currentRatio', np.nan))
    data['Beta'].append(info.get('beta', np.nan))
    data['Price'].append(info.get('regularMarketPrice', np.nan))
    data['52w Low'].append(info.get('fiftyTwoWeekLow', np.nan))
    data['52w High'].append(info.get('fiftyTwoWeekHigh', np.nan))
    # datas.append(data)
data_df = pd.DataFrame(data)
```
